```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load data
customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
transactions = pd.read_csv('Transactions.csv')
```

```python
# Add Year and Month columns to transactions and customers datasets
# For Transactions Dataset
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
transactions['Year'] = transactions['TransactionDate'].dt.year
transactions['Month'] = transactions['TransactionDate'].dt.month

# For Customers Dataset
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
customers['SignupYear'] = customers['SignupDate'].dt.year
customers['SignupMonth'] = customers['SignupDate'].dt.month
```

```python
customers.head(10)
```

```python
import pandas as pd

def explore_dataset(df, date_columns=None):
    """
    Perform initial dataset exploration.

    Parameters:
    df (pd.DataFrame): The dataset to explore.
    date_columns (list): List of column names to be converted to datetime format.

    Returns:
    None
    """
    print("Dataset Head (First 10 Rows):")
    print(df.head(10))
    print("\nDataset Info (Data Types and Non-Null Counts):")
    print(df.info())
    print("\nMissing Values Per Column:")
    print(df.isnull().sum())
    
    # Convert specified columns to datetime
    if date_columns:
        for col in date_columns:
            if col in df.columns:
                df[col] = pd.to_datetime(df[col])
                print(f"\n'{col}' converted to datetime format.")
            else:
                print(f"\nColumn '{col}' not found in the dataset.")
    print("\nPreview of Dataset After Conversions:")
    print(df.head(10))
```

```python
customers.head()
```

```python


# Use the function
explore_dataset(customers, date_columns=['SignupDate'])
```

```python

# Use the function
explore_dataset(transactions, date_columns=['TransactionDate'])
```

```python

# Use the function
explore_dataset(products)
```

## EDA Analysis

# 1. Count of transactions per customer

```python
transactions_per_customer = transactions['CustomerID'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=transactions_per_customer.head(20).index, y=transactions_per_customer.head(20).values, palette='Blues_d')
plt.title("Top 20 Customers by Number of Transactions")
plt.xlabel("CustomerID")
plt.ylabel("Number of Transactions")
for i, v in enumerate(transactions_per_customer.head(20).values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)  # Annotate bars
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# 2. Average transaction amount per product category


```python
avg_value_per_category = transactions.groupby('ProductID')['TotalValue'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_value_per_category.head(10).index, y=avg_value_per_category.head(10).values, palette='Oranges_d')
plt.title("Top 10 Products by Average Transaction Value")
plt.xlabel("ProductID")
plt.ylabel("Average Transaction Value")
for i, v in enumerate(avg_value_per_category.head(10).values):
    plt.text(i, v + 0.5, f"{v:.2f}", ha='center', va='bottom', fontsize=10)  # Annotate bars
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# 3. Customer signup trends over time

```python
signup_trends = customers['SignupYear'].value_counts().sort_index()
plt.figure(figsize=(12, 7))
sns.lineplot(x=signup_trends.index, y=signup_trends.values, marker='o', color='green')
plt.title("Customer Signup Trends Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Signups")
for i, v in enumerate(signup_trends.values):
    plt.text(signup_trends.index[i], v , str(v), ha='center', va='bottom', fontsize=10)  # Annotate points
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# 4. Distribution of transaction values


```python
plt.figure(figsize=(12, 6))
sns.histplot(transactions['TotalValue'], bins=30, kde=True, color='purple')
plt.title("Distribution of Transaction Values")
plt.xlabel("Transaction Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

# 5. Product popularity by sales volume

```python
product_sales_volume = transactions['ProductID'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=product_sales_volume.head(10).index, y=product_sales_volume.head(10).values, color='orange')
plt.title("Top 10 Products by Sales Volume")
plt.xlabel("ProductID")
plt.ylabel("Sales Volume")
for i, v in enumerate(product_sales_volume.head(10).values):
    plt.text(i, v , str(v), ha='center', va='bottom', fontsize=10)  # Annotate bars
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# 6. Total revenue per product

```python
total_revenue_per_product = transactions.groupby('ProductID')['TotalValue'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=total_revenue_per_product.head(10).index, y=total_revenue_per_product.head(10).values, color='skyblue')
plt.title("Top 10 Products by Total Revenue")
plt.xlabel("ProductID")
plt.ylabel("Total Revenue")
for i, v in enumerate(total_revenue_per_product.head(10).values):
    plt.text(i, v + 50, f"{v:.2f}", ha='center', va='bottom', fontsize=10)  # Annotate bars
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# 7. Correlation heatmap (Transactions dataset)


```python
# Select only numeric columns
numeric_columns = transactions.select_dtypes(include=['number']).columns

# Compute the correlation matrix on numeric columns
correlation_matrix = transactions[numeric_columns].corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix (Transactions)")
plt.tight_layout()
plt.show()
```

# 8. Average purchase amount by customer


```python
avg_purchase_by_customer = transactions.groupby('CustomerID')['TotalValue'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
avg_purchase_by_customer.head(10).plot(kind='bar', color='teal')
plt.title("Top 10 Customers by Average Purchase Value")
plt.xlabel("CustomerID")
plt.ylabel("Average Purchase Value")
plt.xticks(rotation=45)
for i, v in enumerate(avg_purchase_by_customer.head(10)):
    plt.text(i, v + 5, f"{v:.2f}", ha='center', va='bottom', fontsize=10)  # Annotate bars
plt.tight_layout()
plt.show()
```

# 9. Frequency of transaction dates


# Month-wise Transaction Frequency

```python
transactions['YearMonth'] = transactions['TransactionDate'].dt.to_period('M')  # Get Year-Month period
transaction_date_freq = transactions['YearMonth'].value_counts().sort_index()
plt.figure(figsize=(14, 6))
sns.lineplot(x=transaction_date_freq.index.astype(str), y=transaction_date_freq.values, marker='o', color='magenta')
plt.title("Transaction Frequency Month-wise")
plt.xlabel("Year-Month")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=90)
for i, v in enumerate(transaction_date_freq.values):
    plt.text(i, v + 5, str(v), ha='center', va='bottom', fontsize=10)  # Annotate points
plt.tight_layout()
plt.show()
```

# 10.Year-wise Transaction Frequency


```python
# Extract the year from the TransactionDate
transactions['Year'] = transactions['TransactionDate'].dt.year

# Calculate the frequency of transactions per year
transaction_year_freq = transactions['Year'].value_counts().sort_index()

# Plot the year-wise transaction frequency
plt.figure(figsize=(14, 6))
sns.lineplot(x=transaction_year_freq.index.astype(str), y=transaction_year_freq.values, marker='o', color='magenta')
plt.title("Transaction Frequency Year-wise")
plt.xlabel("Year")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=90)

# Annotate each point with its value
for i, v in enumerate(transaction_year_freq.values):
    plt.text(i, v + 5, str(v), ha='center', va='bottom', fontsize=10)  # Annotate points

plt.tight_layout()
plt.show()
```

# 11. Pairplot (for numerical relationships) with 'ProductID' as hue and a custom palette


```python
plt.figure(figsize=(12, 12))
sns.pairplot(transactions[['TotalValue', 'Quantity', 'ProductID']], hue='Quantity', palette='Set2')
plt.suptitle("Pairplot for Numerical Variables with Hue", y=1.02)
plt.tight_layout()
plt.show()
```

# 12. Jointplot (to explore relationships between two variables)


```python
sns.jointplot(x='TotalValue', y='Quantity', data=transactions, kind='scatter', color='teal')
plt.suptitle("Jointplot between TotalValue and Quantity", y=1.02)
plt.tight_layout()
plt.show()
```

# 13. Regression plot (for predicting trends)


```python
sns.regplot(x='Quantity', y='TotalValue', data=transactions, scatter_kws={'s': 10}, line_kws={'color': 'red'})
plt.title("Regression Plot between Quantity and Total Value")
plt.tight_layout()
plt.show()
```

# 14. Boxplot (for distribution by product category)


```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='ProductID', y='TotalValue', data=transactions, palette='coolwarm')
plt.title("Boxplot of Total Value by Product Category")
plt.xlabel("ProductID")
plt.ylabel("Total Value")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
